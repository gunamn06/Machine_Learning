```python
import numpy as np
import matplotlib.pyplot as plt

# Data (Hours studied vs Pass/Fail)
X = np.array([1, 2, 3, 4, 5, 6])
y = np.array([0, 0, 0, 1, 1, 1])   # 0 = Fail, 1 = Pass

# Initialize
m = 0
c = 0
lr = 0.1
epochs = 1000
n = len(X)

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Training
for i in range(epochs):
    z = m * X + c
    y_pred = sigmoid(z)

    # Gradients
    dm = (1/n) * np.sum((y_pred - y) * X)
    dc = (1/n) * np.sum(y_pred - y)

    # Update
    m = m - lr * dm
    c = c - lr * dc

# Final predictions
z = m * X + c
y_pred = sigmoid(z)

print("Slope (m):", m)
print("Intercept (c):", c)

# Prediction for new value
x_new = 3.5
prob = sigmoid(m * x_new + c)

print("Probability of Passing:", prob)
print("Final Prediction:", 1 if prob >= 0.5 else 0)

# Plot
plt.figure()
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, label="Sigmoid Curve")
plt.xlabel("Hours Studied")
plt.ylabel("Pass(1)/Fail(0)")
plt.title("Logistic Regression")
plt.legend()
plt.show()
```
