```python
import numpy as np

# Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Initialize
m = 0
c = 0
lr = 0.01
epochs = 1000
n = len(X)

# Gradient Descent
for i in range(epochs):
    y_pred = m * X + c

    dm = (-2/n) * sum(X * (y - y_pred))
    dc = (-2/n) * sum(y - y_pred)

    m = m - lr * dm
    c = c - lr * dc


print("Slope:", m)
print("Intercept:", c)

print("Prediction for x=6:", m*6 + c)
```
